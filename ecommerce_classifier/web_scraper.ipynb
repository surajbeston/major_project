```python
filtered_categories = []
for category in categories["data"]["child"][0]["child_menu"]:
    if category['title'] != 'Wholesale Product':
        filtered_categories.append({'title': category['title'], 'url': category['url']})
```

```python
all_products = []
for category in filtered_categories:
    products = requests.get('https://www.gyapu.com/api/product/category/{}')
    
```

```python
with open('gyapu_new_price.json', 'a') as file:
    all_products = []
    for category in filtered_categories:
        data = requests.get(f'https://www.gyapu.com/api/product{category["url"]}')
        sub_categories = json.loads(data._content)["data"]["children"]
        for sub_category in sub_categories:
            data = requests.get(f"https://www.gyapu.com/api/product/productbycategory/{sub_category['url']}/?size=1000000")
            products = json.loads(data._content)["data"]
            for product in products:
                image = None
                if len(product['image']) > 0:
                    try:
                        image = product['image'][0]['document']['path']
                    except:
                        pass
                price = 0
                if len(product['variant']) > 0:
                    try:
                        price = product['variant'][0]['price']
                    except: 
                        pass
                truncated_product = {'name': product['name'], 'description': product['description'], 'image': image ,'category': category['title'], 'sub_category': sub_category['title'], 'price': price, 'url': 'https://gyapu.com/detail/' + product['url_key']}
                all_products.append(truncated_product)                
        print(len(all_products))
        print(truncated_product)
        print("----------------------")
```
